```python
x_train=train_datagen.flow_from_directory(r'/home/wsuser/work/Dataset Plant Disease/Veg-dataset/test_set',target_size=(256,256),class_mode='categorical',batch_size=32)
x_val=val_datagen.flow_from_directory(r'/home/wsuser/work/Dataset Plant Disease/Veg-dataset/train_set',target_size=(256,256),class_mode='categorical',batch_size=32)
```

```python
x_train.class_indices
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Convolution2D,MaxPooling2D,Flatten
```

```python
model=Sequential()
```

```python
model.add(Convolution2D(32,(3,3),input_shape=(256,256,3),activation='relu'))
```

```python
model.add(MaxPooling2D(pool_size=(2,2)))
```

```python
model.add(Flatten())
```

```python
model.summary()
```

```python
model.add(Dense(300,activation='relu'))
model.add(Dense(150,activation='relu'))
model.add(Dense(6,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit_generator(x_train,steps_per_epoch=len(x_train),validation_data=x_val,validation_steps=len(x_val),epochs=10)
```

```python
model.save('veg.h5')
```
